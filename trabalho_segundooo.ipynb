```python
from ipywidgets import *
# from matplotlib_venn import venn2
from IPython.display import display

# Input Fields
input_a = Text(placeholder='hatama valor A', layout=Layout(width='300px'))
input_b = Text(placeholder='hatama valor B', layout=Layout(width='300px'))
input_c = Text(placeholder='hatama valor C', layout=Layout(width='300px'))

# Output Areas
output_display = Output(layout=Layout(width='250px', height='150px', border='2px solid black', background_color='#00BFFF'))
output_preview = Output(layout=Layout(width='250px', height='150px', border='2px solid black', background_color='#00BFFF'))

# Variabel resultsdu antes
previous_result = {'text': ''}

# Fungsi troka input 
def parse_input(text):
    try:
        return set(map(int, text.split(','))) if text.strip() else set()
    except ValueError:
        return set()

# Fungsi halo tombol operasaun
def make_op_button(label, func):
    btn = Button(description=label, layout=Layout(width='70px'), button_style='success')
    
    def on_click(_):
        a = parse_input(input_a.value)
        b = parse_input(input_b.value)
        c = parse_input(input_c.value)
        result = func(a, b, c)
        result_text = f"{label}:\n{result}"

        # Update preview antes update display
        output_preview.clear_output()
        with output_preview:
            print(previous_result['text'])

        output_display.clear_output()
        with output_display:
            print(result_text)

        previous_result['text'] = result_text  # Save rezultadu iha preview

    btn.on_click(on_click)
    return btn

# Tombol Operasi
ops_top = HBox([
    make_op_button('A ∪ C', lambda a, b, c: a.union(c)),
    make_op_button('A ∪ B', lambda a, b, c: a.union(b)),
    make_op_button('A ∩ B', lambda a, b, c: a.intersection(b)),
    make_op_button('B ∩ C', lambda a, b, c: b.intersection(c)),
])
ops_bottom = HBox([
    make_op_button('A - B', lambda a, b, c: a.difference(b)),
    make_op_button('B - C', lambda a, b, c: b.difference(c)),
    make_op_button('C - B', lambda a, b, c: c.difference(b)),
])

# Tombol Clear
btn_clear = Button(description="Clear", button_style='success')
def clear_all(_):
    # input_a.value = ''
    # input_b.value = ''
    # input_c.value = ''
    output_display.clear_output()
    output_preview.clear_output()
    previous_result['text'] = ''
btn_clear.on_click(clear_all)

# troka HTML header sai widgets.HTML
header = HTML(value="<h2 style='color:white; background-color:#4169E1; padding:10px'>PROGRAM SETS</h2>")

# Layout final
layout = VBox([
    header,
    HBox([Label("A"), input_a]),
    HBox([Label("B"), input_b]),
    HBox([Label("C"), input_c]),
    ops_top,
    ops_bottom,
    HBox([
        VBox([Label("Display"), output_display]),
        VBox([Label("Preview"), output_preview])
    ]),
    btn_clear
])

display(layout)
```
